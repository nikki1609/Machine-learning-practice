```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
y=x**3
plt.plot(x,y)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(x,y,'b')
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('title')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.35,0.4,0.5])
axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig,axes=plt.subplots(nrows=3,ncols=4)
plt.tight_layout()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
for curraxes in axes:
    curraxes.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('first')
axes[1].plot(y,x)
axes[1].set_title('second')
```

```python
fig=plt.figure(figsize=(3,2))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig=plt.figure(figsize=(8,4))
ax=fig.add_axes([0.1,0.1,0.4,0.4])
ax.plot(x,x**2,label='xsqud',color='k')
ax.plot(x,x**3,label='xcube')
ax.legend(loc=0)
```

```python
fig.savefig('my_picture.png',dpi=200)
```

```python
fig=plt.figure(figsize=(8,4))
ax=fig.add_axes([0.1,0.1,0.4,0.4])
ax.plot(x,x**2,linewidth=5,color='red')
```

```python
fig=plt.figure(figsize=(12,8))       
ax=fig.add_axes([0.1,0.1,0.4,0.4])
ax.plot(x,x**2,lw=4,alpha=50,color='red',linestyle='-',marker='o',#marker_symbols=('$','*',',','.','o','+',1,2,3,4)
        markersize=10,markerfacecolor='yellow',markeredgecolor='green')  
```
